Получим тесты, которые работают

```python
path_dir = 'json_analytics_from_test\\suite_windows_test_before_fix'
```

```python
from ParseOutputPytest import ParseOutputPytest
pop = ParseOutputPytest()
```

```python
# Get a list of tests in the folder
import os
list_files = list(os.walk(path_dir))[0][2]
```

```python
# Get for each file - dict
list_test = []
for name_file_test in list_files:
    new_dict = pop.get_dict_test_duration(f'{path_dir}\\{name_file_test}')
    for key in new_dict.keys():
        if '<lambda>' not in key and ' ' not in key:
            list_test.append(key)
```

```python
tup_test = tuple(sorted(list_test))
print(f'Passed test: {len(tup_test)}')
```

Запустим тесты по группам

```python
n = 20  # блоки по 20 тестов
chunks = tuple(tup_test[i:(i + n)] for i in range(0, len(tup_test), n))
```

```python
environment_variables = 'set MODIN_EXPERIMENTAL=1 && set MODIN_ENGINE=Python'
args_pytest = '--simulate-cloud=normal'
path_dir = 'C:\\prog\\modin'
path_to_save = 'test_analytics\\json_analytics_from_test\\passed_test'
```

```python
for i, tuple_names in enumerate(chunks):
#     if i < 40:
#         continue
#     print(tuple_names)
    str_test = ' '.join(tuple_names)
    pop.start_test(path_dir, str_test, environment_variables=environment_variables, 
            args_pytest=args_pytest,
            path_dir_for_save=path_to_save,
            new_name=f'passed_test_{i}')
```
